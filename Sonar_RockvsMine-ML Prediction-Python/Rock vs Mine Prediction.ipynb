## Training and Test the data

<br>

```python
X_train,X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1, stratify= y , random_state = 1)
```

```python
print("\n","X shape = ", X.shape,"\n", "X_train shape = ", X_train.shape,"\n", "X_test shape", X_test.shape,"\n")
```

## Model Training

<br>

___Logistic Regression___

```python
model = LogisticRegression()
model
```

```python
model.fit(X_train, y_train)
```

## Model Evaluation

- __Accuracy of training data ;__

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(y_train, X_train_prediction)
```

```python
print("Accuracy of training data is : " , "%",round(training_data_accuracy,2))
```

```python
sm.confusion_matrix(y_train, X_train_prediction)
```
